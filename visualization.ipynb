```python
import configparser
import pickle
import pandas as pd
from IPython.core.display import display
from plotly.graph_objs import Layout

config = configparser.ConfigParser()
config.read('config/appconfig.ini')
hourly_filename = '2018-03-22_to_2020-06-02_hourly_weather_data.pkl'
daily_filename = '2018-03-22_to_2020-06-02_daily_weather_data.pkl'

file = open(config['DATA_DIRECTORIES']['processed_weather_data_directory'] + '\\' + hourly_filename, 'rb')
loaded_hourly_data = pickle.load(file)

file = open(config['DATA_DIRECTORIES']['processed_weather_data_directory'] + '\\' + daily_filename, 'rb')
loaded_daily_data = pickle.load(file)
```

```python
# Display loaded hourly weather dataframe
display(loaded_hourly_data)
```

```python
# Display loaded daily weather dataframe
display(loaded_daily_data)
```

```python
# Display temperatures over the time period
import plotly.express as px
fig = px.line(loaded_hourly_data, x=loaded_hourly_data.time, y=loaded_hourly_data.temperature)
fig.update_layout(xaxis_title="Month, Year", yaxis_title="Temperature [F]", title="DC House Historical Weather Data",
                  template="none")
fig.show()
```

```python
# Display daily weather data
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=loaded_daily_data.time, y=loaded_daily_data.temperatureHigh, name="Daily High Temperature"))
fig.add_trace(go.Scatter(x=loaded_daily_data.time, y=loaded_daily_data.temperatureLow, name="Daily Low Temperature"))
fig.add_trace(go.Scatter(x=loaded_daily_data.time, y=loaded_daily_data.dewPoint, name="Daily Average Dew Point"))
fig.update_layout(xaxis_title="Month, Year", yaxis_title="Temperature [F]", title="DC House Historical Weather Data",
                  template="none", legend=dict(
                                    orientation="h",
                                    yanchor="bottom",
                                    y=1.02,
                                    xanchor="right",
                                    x=1.02))
fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=loaded_daily_data.time, y=loaded_daily_data.humidity, name="Daily Average Humidity"))
fig.update_layout(xaxis_title="Month, Year", yaxis_title="Relative Humidity [%]", title="DC House Historical Weather "
                  "Data", template="none", legend=dict(
                                            orientation="h",
                                            yanchor="bottom",
                                            y=1.02,
                                            xanchor="right",
                                            x=1))
fig.show()
```
